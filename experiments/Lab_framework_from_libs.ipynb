## STS Analyzer

Generic class to evaluate and analyze STS results. During the labs, similarity was measured using predefined functions. In this case, what we provide is a model that can be used to measure similarity.

```python
sts = STSAnalyzer(preprocess_fun=spacy_preprocess_reviews, 
                  model=BaselineModel(strategy="constant", constant=0), 
                  name='Baseline (Constant 0)')

sts.load_data(ls_files=STSAnalyzer.LS_FILES_TEST)
sts.get_results()
sts.plot_col_analysis()
```

```python
sts2 = STSAnalyzer(preprocess_fun=spacy_preprocess_reviews, 
                  model=BaselineModel(strategy="constant", constant=3), 
                  name='Baseline (constant 3)')

sts2.load_data(ls_files=STSAnalyzer.LS_FILES_TEST)
sts2.get_results()
sts2.plot_col_analysis([sts])
```

```python
sts3 = STSAnalyzer(preprocess_fun=spacy_preprocess_reviews, 
                  model=BaselineModel(strategy="constant", constant=5), 
                  name='Baseline (constant 5)')

sts3.load_data(ls_files=STSAnalyzer.LS_FILES_TEST)
sts3.get_results()
sts3.plot_col_analysis([sts, sts2])
```

```python
sts4 = STSAnalyzer(preprocess_fun=spacy_preprocess_reviews, 
                  model=BaselineModel(strategy="random", constant=5), 
                  name='Baseline (random)')

sts4.load_data(ls_files=STSAnalyzer.LS_FILES_TEST)
sts4.get_results()
sts4.plot_col_analysis([sts, sts2, sts3])
```
